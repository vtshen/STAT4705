# Demos on Distributions

```python
from scipy.stats import uniform
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

## Continuous Uniform Distributions

### Density function

```python

#Let's set an A and B
A=0
B=4
# Set x as 100 linearly spaced points between A and B
x = np.linspace(A,B,100)

# Use uniform(loc=start point,scale=endpoint)
rv = uniform(A,B)

#Plot the PDF of that uniform distirbution
plt.plot(x,rv.pdf(x))
plt.show()
```

```python
# CDF curve
plt.plot(x,rv.cdf(x))
plt.show()
```

```python
# CDF value
rv.cdf(1.5)
```

## Normal Distribution
#### Density function

```python
from IPython.display import Image
Image(url='http://upload.wikimedia.org/wikipedia/commons/thumb/2/25/The_Normal_Distribution.svg/725px-The_Normal_Distribution.svg.png')
```

```python
mean = 0
std = 1

X = np.arange(-4,9,0.01)

#Create the normal distribution for the range
Y = stats.norm.pdf(X,mean,std)
plt.plot(X,Y)
```

```python
# CDF curve
mean = 0
std = 1
X = np.arange(-4,4,0.01)

Z = stats.norm.cdf(X,mean,std)
plt.plot(X,Z)
```

```python
mean = 0
std = 1
cutoff = 1.96
stats.norm.cdf(cutoff,mean,std)
```

Q: how to calculate the probability by integral of the density function?

1. use CDF
2. use integral directly, thanks to the suggestion from 

```python
# this is method 2, direct integral
from scipy.integrate import quad
def densityFun(x):
    # standard normal density function, N(0,1)
    return (1/(2 * np.pi)**0.5) * np.exp(-x**2/2)

ans, err = quad(densityFun, 0, 1)
print (ans)
print(err)
```

```python
# another calculation
ans, err = quad(densityFun, 0, np.inf)
print (ans)
print(err)
```
