# Continuous Distribution
## Continuous Uniform Distributions

### Density function

```python

#Let's set an A and B
A=0
B=4
# Set x as 100 linearly spaced points between A and B
x = np.linspace(A,B,100)

# Use uniform(loc=start point,scale=endpoint)
rv = uniform(A,B)

#Plot the PDF of that uniform distirbution
plt.plot(x,rv.pdf(x))
plt.show()
```

```python
# CDF curve
plt.plot(x,rv.cdf(x))
plt.show()
```

```python
# CDF value
rv.cdf(1.5)
```

Q: how to calculate the probability by integral of the density function?

1. use CDF
2. use integral directly, thanks to the suggestion from 

```python
# this is method 2, direct integral
def densityNormal(x, mu = 1, var = 4):
    # standard normal density function, N(0,1)
    return (1/(2 * np.pi * var)**0.5) * np.exp(-(x-mu)**2/(2*var))

ans= quad(densityNormal, 3, np.inf)
print (ans)
```

```python
# another calculation
ans = quad(densityNormal, 0, np.inf)
print (ans)
```

### Chi-Square distributions
the density function


```python
x = np.linspace(0, 10, 1000)

linestyles = [':', '--', '-.', '-']
deg_of_freedom = [1, 2, 3, 4]
for df, ls in zip(deg_of_freedom, linestyles):
  plt.plot(x, stats.chi2.pdf(x, df), linestyle=ls, label=r'$df=%i$' % df)

# figure annotation
plt.xlim(-2, 10)
plt.ylim(0, 0.5)
plt.xlabel('$\chi^2$')
plt.ylabel(r'$f(\chi^2)$')
plt.title(r'$\chi^2\ \mathrm{Distribution}$')
plt.legend()
```

```python
# cdf function
x = np.linspace(0, 10, 1000)

linestyles = [':', '--', '-.', '-']
deg_of_freedom = [1, 2, 3, 4]
for df, ls in zip(deg_of_freedom, linestyles):
  plt.plot(x, stats.chi2.cdf(x, df), linestyle=ls, label=r'$df=%i$' % df)

# figure annotation
plt.xlabel('$\chi^2$')
plt.ylabel(r'$f(\chi^2)$')
plt.title(r'$\chi^2\ \mathrm{Distribution}$')
plt.legend()
```

### F distribution
the density function

```python
x = np.linspace(-1, 6, 1000)

linestyles = [':', '--', '-.', '-']
d1_values = [1, 5, 2, 10]
d2_values = [1, 2, 5, 50]
for d1, d2, ls in zip(d1_values, d2_values, linestyles):
  plt.plot(x, stats.f.pdf(x, d1, d2, 0), linestyle=ls,
           label=r'$d_1=%i,\ d_2=%i$' % (d1, d2))

# figure annotation
plt.xlim(-1, 6)
plt.ylim(0, 1.5)
plt.xlabel('$x$')
plt.ylabel(r'$f(x)$')
plt.title(r'$Fishers\ \mathrm{Distribution}$')
plt.legend()
```

```python
x = np.linspace(-1, 6, 1000)

linestyles = [':', '--', '-.', '-']
d1_values = [1, 5, 2, 10]
d2_values = [1, 2, 5, 50]
for d1, d2, ls in zip(d1_values, d2_values, linestyles):
  plt.plot(x, stats.f.cdf(x, d1, d2, 0), linestyle=ls,
           label=r'$d_1=%i,\ d_2=%i$' % (d1, d2))

# figure annotation
plt.xlim(-1, 6)
plt.ylim(0, 1.5)
plt.xlabel('$x$')
plt.ylabel(r'$f(x)$')
plt.title(r'$Fishers\ \mathrm{CDF Distribution}$')
plt.legend()
```

### Geometric Distribution


```python
# plot the pmf
prob = 0.3
x=np.arange(0, 30)
pmf_geom = geom.pmf(x,prob)

import seaborn as sns
plt.plot(x,pmf_geom, "o")
plt.xlim(-5, 30)
```

```python
# Geometric cdf plot
prob = 0.3
x=np.arange(0, 30)
cdf_geom = geom.cdf(x,prob)
plt.plot(x,cdf_geom, linestyle='steps-mid')
```

```python
# calculate pmf in geometric distribution
prob = 0.3
x = 2
geom.pmf(x, prob)
```

### Poisson Distribution

```python
# plot the pmf
lamb = 10
x=np.arange(0, 30)
pmf_pois = poisson.pmf(x,lamb)

import seaborn as sns
plt.plot(x,pmf_pois, "o")
```

```python
# Poisson cdf plot
cdf_pois = poisson.cdf(x,lamb)
plt.plot(x,cdf_pois, linestyle='steps-mid')
```

```python
# calculate the probability
lamb = 10
k = 5
prob_up_to_number_k = poisson.cdf(k,lamb)
print(prob_up_to_number_k)
```

```python
# this is method 2, direct sum
def densityPoisson(x, lamb = 10):
    # poisson probability mass function, N(0,1)
    return (np.exp(-lamb)*lamb**x/ math.factorial(x))

print(densityPoisson(3))
```
