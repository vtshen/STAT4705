```python
from scipy.stats import uniform
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.integrate import quad
from scipy.stats import poisson
import math
```

# Continuous Distribution
## Continuous Uniform Distributions

### Density function

```python

#Let's set an A and B
A=0
B=4
# Set x as 100 linearly spaced points between A and B
x = np.linspace(A,B,100)

# Use uniform(loc=start point,scale=endpoint)
rv = uniform(A,B)

#Plot the PDF of that uniform distirbution
plt.plot(x,rv.pdf(x))
plt.show()
```

```python
# CDF curve
plt.plot(x,rv.cdf(x))
plt.show()
```

```python
# CDF value
rv.cdf(1.5)
```

```python
mean = 0
std = 1

X = np.arange(-10,10,0.01)

#Create the normal distribution for the range
Y = stats.norm.pdf(X,mean,std)
plt.plot(X,Y)
```

```python
# CDF curve
mean = 0
std = 1
X = np.arange(-10,10,0.01)

Z = stats.norm.cdf(X,mean,std)
plt.plot(X,Z)
```

```python
mean = 0
std = 1
cutoff = 1.96
stats.norm.cdf(cutoff,mean,std)
```

Q: how to calculate the probability by integral of the density function?

1. use CDF
2. use integral directly, thanks to the suggestion from 

```python
# this is method 2, direct integral
def densityNormal(x, mu = 1, var = 4):
    # standard normal density function, N(0,1)
    return (1/(2 * np.pi * var)**0.5) * np.exp(-(x-mu)**2/(2*var))

ans= quad(densityNormal, 3, np.inf)
print (ans)
```

```python
# another calculation
ans = quad(densityNormal, 0, np.inf)
print (ans)
```

# Discrete Distribution
## Poisson Distribution

```python
# plot the pmf
lamb = 10
x=np.arange(0, 30)
pmf_pois = poisson.pmf(x,lamb)

import seaborn as sns
plt.plot(x,pmf_pois, "o")
```

```python
# Poisson cdf plot
cdf_pois = poisson.cdf(x,lamb)
plt.plot(x,cdf_pois, linestyle='steps-mid')
```

```python
# calculate the probability
lamb = 10
k = 5
prob_up_to_number_k = poisson.cdf(k,lamb)
print(prob_up_to_number_k)
```

```python
# this is method 2, direct sum
def densityPoisson(x, lamb = 10):
    # poisson probability mass function, N(0,1)
    return (np.exp(-lamb)*lamb**x/ math.factorial(x))

print(densityPoisson(3))
```
